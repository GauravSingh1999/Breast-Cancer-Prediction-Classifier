Importing Libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing Dataset

```
import pandas as pd
df = pd.read_csv('Breast_Cancer.csv')
print("Importing Dataset")
df.head()
```

Matrices Of Features

```
print("Creating Matrix Of Features")
X = df.iloc[:, 1].values
y = df.iloc[:, 2].values
```

Labelling

```
print("Labelling of Dependent data as 0 and 1")
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X = le.fit_transform(X)
```

Reshaping Shaping Size Of X- Matrix

```
X = X.reshape(-1, 1)
print(X)
```

Reshaping Size Of y-Matrix 

```
y = y.reshape(-1, 1)
print(y)
```

Feature Scaling:

```
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```
print(X)
```

```
print(y)
```

Training the SVR model on the whole dataset

```
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X, y)
```

Predicting a new result for Recurrent('R' as 1)

```
sc_y.inverse_transform(regressor.predict(sc_X.transform([[1]])))
```

Visualisation

```
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'red')
plt.plot(sc_X.inverse_transform(X), sc_y.inverse_transform(regressor.predict(X)), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Outcome')
plt.ylabel('Recurrence Time')
plt.show()
```
